```python
import os

# folder saved_samples/DAMH_accepted

folder = 'saved_samples/config_mcmc_bayes/DAMH_accepted'
# read all files in the folder into numpy arrays
files = os.listdir(folder)
files.sort()
files = files[1:]
print(files)
```

```python
# load all files from files into numpy arras as csv files
import numpy as np
all_files = []
for file in files:
    with open(folder + '/' + file, 'r') as f:
        all_files.append(np.loadtxt(f, delimiter=','))
```

```python
# concatenate all arrays into one by rows
all_files = np.concatenate(all_files, axis=0)
```

```python
# compute the absolte value of difference between  columns 1 and 2
alpha = np.exp(- np.abs(all_files[:, 1] - all_files[:, 2]))
```

```python
# plot sorted alpha
import matplotlib.pyplot as plt
plt.plot(np.sort(alpha))
plt.show()
```
